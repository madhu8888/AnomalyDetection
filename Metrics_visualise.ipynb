```python
import numpy as np
import pandas as pd
import import_ipynb
import Data_generation as Data_generate
import Algo_evaluate as algo
```

```python
d_set = ['Train', 'Test']
eval_met = ['Pre', 'Rec', 'Acc', 'F1']
cont_factor = np.arange(0.05,0.55,0.05)
```

```python
def column_list(d_set, eval_met ):
  column_names = [];
  for j in d_set:
    for k in eval_met:
      column_names.append(j+"_"+k)
  return column_names;
```

```python
column_names = column_list(d_set, eval_met);
```

# Vary contamination factor

```python
from pyod.models.iforest import IForest
from pyod.models.ocsvm import OCSVM
from pyod.models.pca import PCA
from pyod.models.lof import LOF
```

```python
def detect_outliers( cont_factor, clf, df_return ):
  tr_acc_lst =[];
  tr_pre_lst =[];
  tr_rec_lst =[];
  tr_f1_lst =[];
  tst_acc_lst =[];
  tst_pre_lst =[];
  tst_rec_lst =[];
  tst_f1_lst =[];
  for i in cont_factor:
    df = Data_generate.GenerateDataset(n_Gauss_comp=1, samples=100, cont=i, n_features=5 , num_cat=0 , missing_percent=0, cov_type="tied-sph", dep_per=0 , uni_low=10, uni_up=20, mu_low_range=20,mu_high_range=40)
    clf.contamination = i;
    train_accuracy , train_precision, train_recall, train_f1_score , test_accuracy , test_precision, test_recall, test_f1_score = algo.algo_metric_eval(df,clf );


    tr_acc_lst.append(train_accuracy);
    tr_pre_lst.append(train_precision);
    tr_rec_lst.append(train_recall);
    tr_f1_lst.append(train_f1_score);
    tst_acc_lst.append(test_accuracy);
    tst_pre_lst.append(test_precision);
    tst_rec_lst.append(test_recall);
    tst_f1_lst.append(test_f1_score);
						
  df_return['Train_Pre'] = tr_pre_lst;
  df_return['Train_Rec'] = tr_rec_lst;
  df_return['Train_Acc'] = tr_acc_lst;
  df_return['Train_F1'] = tr_f1_lst;

  df_return['Test_Pre'] = tst_pre_lst;
  df_return['Test_Rec'] = tst_rec_lst;
  df_return['Test_Acc'] = tst_acc_lst;
  df_return['Test_F1'] = tst_f1_lst;

  #return df_return
```

```python
def plot_metrics( x_lst, y_lst, x_axis , y_axis , x_lbl_lst , plot_title):
  plt.title(plot_title);
  plt.xlabel(x_axis);
  plt.ylabel(y_axis);
  for i in range(len(x_lst)):
    plt.plot( y_lst, x_lst[i] , label = x_lbl_lst[i])

  plt.legend()
  plt.show()
```
